```python
# Clean and standardize the 'country' column
def clean_country(val):
    if pd.isnull(val):
        return 'Unknown'

    # Remove brackets and quotes, then split by comma if multiple
    cleaned = val.replace('[', '').replace(']', '').replace("'", "").strip()

    # Take the first country if multiple are listed
    first_country = cleaned.split(',')[0].strip()

    return first_country.upper() if first_country else 'Unknown'

# Apply the function to the country column
amazon['production_countries'] = amazon['production_countries'].apply(clean_country)

# View unique cleaned values (optional)
print(amazon['production_countries'].value_counts().head(10))
```

```python

amazon_exploded = amazon.explode('genres')
genre_scores = amazon_exploded.groupby('genres')['imdb_score'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=genre_scores.values, y=genre_scores.index, palette='rainbow')
plt.title('Top 10 Genres by Average IMDb Score')
plt.xlabel('Average IMDb Score')
plt.ylabel('Genre')

for i, score in enumerate(genre_scores.values):
    ax.text(score + 0.05, i, f"{score:.2f}", va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Create runtime buckets
def categorize_runtime(minutes):
    if minutes < 30:
        return 'Short (<30 min)'
    elif 30 <= minutes <= 60:
        return 'Medium (30–60 min)'
    elif 60 < minutes <= 120:
        return 'Long (60–120 min)'
    else:
        return 'Very Long (>120 min)'

amazon['runtime_category'] = amazon['runtime'].apply(categorize_runtime)

# Plot distribution
plt.figure(figsize=(8,6))
sns.countplot(data=amazon, x='runtime_category', order=[
    'Short (<30 min)', 'Medium (30–60 min)', 'Long (60–120 min)', 'Very Long (>120 min)'
], palette='coolwarm')
plt.title('Distribution of Content by Runtime Category')
plt.xlabel('Runtime Category')
plt.ylabel('Number of Titles')
plt.xticks(rotation=15)
plt.show()
```

# **Objective-Expanding global reach by understanding regional content gaps and preferences**

```python
# Prepare data
df = amazon.dropna(subset=['genres', 'production_countries']).copy()
df['genres'] = df['genres'].str.split(', ')
df['production_countries'] = df['production_countries'].str.split(', ')
df = df.explode('genres').explode('production_countries')

# Filter top countries and genres
top_countries = df['production_countries'].value_counts().head(5).index
top_genres = df['genres'].value_counts().head(8).index
filtered = df[df['production_countries'].isin(top_countries) & df['genres'].isin(top_genres)]

# Create pivot table
heatmap_data = filtered.pivot_table(index='genres', columns='production_countries', aggfunc='size', fill_value=0)

# Plot
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title('🎬 Genre Distribution Across Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

# **Why use this Chart ?**
It lets you compare genres across multiple countries at once, making patterns and outliers visually obvious.
# **Insights**

1.Drama genre is the most popular genre across all countries.

2.Comedy is the 2nd most popular genre.

3.Thriller and horror are also going strong showing strong but niche market.

4.Sci-fi is relatively less popular.
